My experiment sought to compare the get() and update() functions for the dictionary data-type. The idea was to randomly select an item from dictionaries of increasing size (max of 10000), and time how fast it takes the computer to retrieve or update that item. As you can see above, the response time stays static regardless of the number of elements contained in the dictionary. This shows that both the get() and update() functions can be considered to be O(1).
